Для регрессии CC50 лучше всего продемонстрировал себя также алгоритм случайного леса.

```python
param_dist_cc = {
    'n_estimators': randint(50, 300),
    'max_depth': randint(3, 10),
    'learning_rate': uniform(0.01, 0.3),
}

random_search_cc = RandomizedSearchCV(
    estimator=GradientBoostingRegressor(),
    param_distributions=param_dist_cc,
    n_iter=50,       
    cv=5,
    scoring='r2',
    n_jobs=-1,
    random_state=42
)

random_search_cc.fit(X_train_cc, y_train_cc)

print("Лучшие параметры:", random_search_cc.best_params_)
```

```python
grad_for_cc = GradientBoostingRegressor(
    n_estimators = random_search_cc.best_params_['n_estimators'],
    max_depth = random_search_cc.best_params_['max_depth'],
    learning_rate = random_search_cc.best_params_['learning_rate'],
    random_state=42
)
grad_for_cc.fit(X_train_cc, y_train_cc)
```

```python
grad_boost_for_cc = grad_for_cc.predict(X_test_cc)

r2_test_grad_for_cc = r2_score(y_test_cc, grad_boost_for_cc)
print("R2 на тесте:", r2_test_grad_for_cc)
```
